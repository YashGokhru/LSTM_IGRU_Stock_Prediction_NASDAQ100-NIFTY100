```python
#Calculating ARR
import numpy as np

portfolio_returns_everyday = np.array(portfolio_returns_everyday)  # Example daily returns

# Calculate cumulative return
cumulative_return = np.prod(1 + portfolio_returns_everyday) - 1

# Number of trading days in the dataset
N_days = len(portfolio_returns_everyday)

#1. ARR (Annualized return ratio)
ARR = (1 + cumulative_return) ** (252 / N_days) - 1

print("Total no. of trading days :", N_days)
print(f"Cumulative Return: {cumulative_return:.4f}")
print(f"Annualized Rate of Return (ARR): {ARR:.4f}")
```

```python
# 2. AVol (Annualized Volatility)
daily_volatility = np.std(portfolio_returns_everyday)
AVol = daily_volatility * np.sqrt(252)
print(f"Annualized Volatility (AVol): {AVol:.4f}")
```

```python
# 3. MDD (Maximum Drawdown)
cumulative_returns = np.cumprod(1 + portfolio_returns_everyday) - 1
cumulative_peaks = np.maximum.accumulate(cumulative_returns)
drawdowns = (cumulative_peaks - cumulative_returns)
MDD = np.max(drawdowns)
print(f"Maximum Drawdown (MDD): {-1*MDD:.4f}")
```

```python
# 4. ASR (Annualized Sharpe Ratio)
risk_free_rate = 0.04 / 252  # Daily risk-free rate (4% annualized)
average_excess_return = np.mean(portfolio_returns_everyday - risk_free_rate)
ASR = average_excess_return / daily_volatility * np.sqrt(252)
print(f"Annualized Sharpe Ratio (ASR): {ASR:.4f}")
```

```python
# 5. CR (Calmar Ratio)
CR = ARR / MDD if MDD != 0 else np.nan
print(f"Calmar Ratio (CR): {CR:.4f}")
```
